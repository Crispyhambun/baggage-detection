```python
def find_unique_labels(dataset):
    """
    Finds and displays the unique labels in the dataset.
    """
    all_labels = []
    for _, target in dataset:
        all_labels.extend(target["labels"].tolist())
    
    unique_labels = set(all_labels)  # Use set to get unique labels
    print(f"Unique Labels: {sorted(unique_labels)}")
    return sorted(unique_labels)

# Example Usage
unique_labels_train = find_unique_labels(train_dataset)
unique_labels_valid = find_unique_labels(valid_dataset)
unique_labels_test = find_unique_labels(test_dataset)
```
